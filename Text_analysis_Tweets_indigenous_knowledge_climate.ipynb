# Top 10 Hashtags used

```python
hashtag = df[[ 'Tweet', 'hashtags', 'url']]

#remove all rows with "None" value in the column 'hashtags'
hashtag =hashtag.dropna(subset=['hashtags'])
```

```python
# new column with the number of Hashtags
hashtag['hashtagCount'] = hashtag['hashtags'].apply(lambda x: len(str(x).split(' ')))
```

```python
hashtag.head(5)
```

```python
hashtag.hashtags	 = hashtag.hashtags.astype('str') 
```

```python
hashtag_list = []                          #CREATE EMPTY LIST 
for i in hashtag.hashtags:    #LOOP OVER EVERY CELL IN ENTITIES_HASHTAGS
   #if pd.notnull(i):                      #IF CELL NOT EMPTY
        tags = i.split()                   #SPLIT EACH CELL INTO SEPARATE HASHTAGS
        for t in tags:                     #FOR EACH TAG IN THE CELL
            t = "#"+t                      #ADD '#' SYMBOL TO BEGINNING OF EACH TAG
            t = t.replace(',', '')         #REMOVE COMMAS FROM END OF TAGS
            t = t.replace('[', '')         #REMOVE [ FROM BEGGING OF TAGS
            t = t.replace(']', '')
            t = t.replace("'", '')
            t = t.lower()                  #MAKE TAG LOWER CASE
            hashtag_list.append(t)         #ADD TAG TO OUR LIST
  
```

```python
print (len(hashtag_list))                    #PRINT NUMBER OF ITEMS IN OUR LIST
print ("--------------------")
hashtag_list[:20]  
```

```python
# let us look most used words for each part of speech
from collections import defaultdict, Counter
```

```python
# 20 most common tokens
HashtagFreq = Counter(hashtag_list)
commonHashtag = HashtagFreq.most_common(10)
```

```python
commonHashtag 
```

```python
Top10Hashtag  = pd.DataFrame (commonHashtag , columns = ['Hashtag', 'Count'])
Top10Hashtag.index  = np.arange(1, len(top10_likes)+1)
Top10Hashtag.style
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = px.bar(Top10Hashtag, x='Hashtag', y="Count", orientation='v', template='plotly_white', title = 'Top 10 Hashtags Used In Tweets',
             height=500, width = 700 
            )

fig.update_yaxes(title_text='Number of Hashtags')
fig.update_xaxes(title_text=' ')
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()

# marker={'color': x, colorscale: 'Viridis'}
```

## Top 10 mentions

```python
df.columns
```

```python
mentions = df[[ 'Tweet', 'mentions', 'url']] 

#remove all rows with "None" value in the column 'hashtags'
mentions  =mentions .dropna(subset=['mentions'])
```

```python
mentions.head(2)
```

```python
mentions.dtypes
```

```python
mentions ['mentions']	 = mentions ['mentions'].astype('str') 
```

 Library

```python
import re
```

```python
def pick_profile(text):
    result = re.findall(r"username='(.*?)'", text)
    return result

mentions['filter_mention']= mentions ['mentions'].apply(lambda x : pick_profile(x))
```

```python
mentions.head(3)
```

```python
mentions['filter_mention']	 = mentions['filter_mention'].astype('str') 
```

```python
mention_list = []                          #CREATE EMPTY LIST 
for i in mentions['filter_mention']:    #LOOP OVER EVERY CELL IN ENTITIES_HASHTAGS
   #if pd.notnull(i):                      #IF CELL NOT EMPTY
        tags = i.split()                   #SPLIT EACH CELL INTO SEPARATE HASHTAGS
        for t in tags:                     #FOR EACH TAG IN THE CELL
            t = "@"+t                      #ADD '#' SYMBOL TO BEGINNING OF EACH TAG
            t = t.replace(',', '')         #REMOVE COMMAS FROM END OF TAGS
            t = t.replace('[', '')         #REMOVE [ FROM BEGGING OF TAGS
            t = t.replace(']', '')
            t = t.replace("'", '')
            t = t.lower()                  #MAKE TAG LOWER CASE
            mention_list.append(t)         #ADD TAG TO OUR LIST
  
```

```python
print (len(mention_list))                    #PRINT NUMBER OF ITEMS IN OUR LIST
print ("--------------------")
mention_list[:10]  
```

```python
# 10 most common tokens
mentionFreq = Counter(mention_list)
commonMention = mentionFreq .most_common(10)
```

```python
commonMention 
```

```python
Top10Mention  = pd.DataFrame (commonMention  , columns = ['Mention', 'Count'])
Top10Mention.index  = np.arange(1, len(top10_likes)+1)
Top10Mention.style
```

```python
import plotly.express as px
import plotly.graph_objects as go
```
